```python
import numpy as np
import style
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
class LinearRegression:
    
    def __init__(self):
        pass
    def fit(self, x,y):
        self.x=np.array(x)
        self.y=np.array(y)
        #m=((((np.mean(self.x)) * (np.mean(self.y))) - (np.mean((self.x) * (self.y)))) / ((np.mean(self.x)**2) - (np.mean(self.x**2))))
        m=np.sum()
        b=np.mean(y)-(m*np.mean(x))
        print(m,b)
        return m,b
    def predict(self,x):
        m,b=self.fit(x,self.y)
        return m*x+b
```

```python
bias=100
x,y,coef = make_regression(n_features=1,bias=bias,noise = 10,random_state=42,coef=True)
```

```python
model=LinearRegression()
model.fit(x,y)
y_line=coef*x+bias
y_pred=model.predict(x)
plt.scatter(x,y)
plt.plot(x,y_line,'g', label='original')
plt.plot(x,y_pred,'r', label='predicted')
plt.legend()
plt.show()
```
