```python
import numpy as np
import matplotlib.pyplot as plt
import style
from sklearn.datasets import make_regression
```

```python
plt.style.use('fivethirtyeight')
```

```python
x,y,coef=make_regression(n_samples=100, n_features=1,bias=100,noise=10, random_state=42, coef=True)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
bias=100
y_new=coef*x+bias
plt.scatter(x,y)
plt.plot(x,y_new,'g')
plt.show()
```

```python
from sklearn import linear_model
```

```python
clf=linear_model.LinearRegression()
clf.fit(x,y)
```

```python
y_predict=clf.predict(x)
```

```python
bias=100
y_new=coef*x+bias
plt.scatter(x,y)
plt.plot(x,y_new,'g', label='original')
plt.plot(x,y_predict, 'r',label='predicted')
plt.legend()
plt.show()
```
